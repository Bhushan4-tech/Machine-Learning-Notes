# K Nearest Neighbour

1. K-Nearest Neighbour is one of the simplest Machine Learning algorithms based on Supervised Learning technique.
2. K-NN algorithm can be used for Regression as well as for Classification but mostly it is used for the Classification
   problem.
3. K-NN is a non-parametric algorithm, which means it does not make any assumption on underlying/given data, whereas 
   it is distance based algorithm.  
4. It Find out dist between new/testing data point and training data points and Select k no. of nearest data point based
   on sortest distance. 
5. To get the predicted value Voting Classifier used in CLassification and Mean/average of k datapoints used in   
   Regression.

# Why KNN is a lazy learner algorithm

* It is also called a lazy learner algorithm because it does not learn from the training set immediately instead 
  it stores the training dataset and at the time of classification, it performs action on the testing dataset.

# Distances in KNN

* As it is a distance based algorithm, it uses following two types of distances:
  1. Euclidean Distance(P=2):
    * E.D. = sqrt((x2-x1)^2 + (y2-y1)^2)
    * where, (x1,y1) and (x2,y2) are the coordinates of testing data point and training datapoints.
  2. Manhatten Distance(P=1):
    * M.D. = |x2-x1| + |y2-y1|
    * Manhatten distance is always greater than Euclidean distance.
* In KNN algorithm we generally uses Euclidean distance rather than Mahnatten distance.

# Backend Working of KNN algorithm

## KNN - Classification

1. In KNN classification type of algorithms the output variables have Categorical/Descrete data.
2. The Target datapoint will calculate distance of itself from each training datapoints i.e. Euclidean Distance.
3. Then Model will select K closest datapoints from all the distances calculated. For that, model will calculate
   distance and sort it in ascending order and select closet K no of datapoints. where, K is the no. of Neighbours.
4. Then "Voting Classifier" is used to decide the predicted class of testing datapoint. i.e. Voting classifier will
   select the class whose maximum data points are in the nearest neibhours.
   * Eg. If there is three classes A, B and C in target variable. The datapoints of each classes in nearest
     neighbours are (for K = 5):
           A = 1
           B = 3
           C = 1
     Now, the B class has maximum datapoint in nearest neighbours, So the predicted class for that
     target varibale will be B-class.
   * If in case there is "Tie" in no. of datapoint then voting classifier will select the class for which maximum
     no. of datapoints are trained.

## KNN - Regression

1. KNN algorithm also used for Regression problems i.e. when the target variable is conitnuous in nature.
2. In this algorithm model will calculate distances of test datapoint from all training datapoints.
3. Then model will select K no of nearest neighbours based on sorted distance.
4. Now the model will select target value by calculating mean/average of Y_train values of all nearest
   neighbours.
   * Eg. For K = 6,
         Y-train values of Nearest neighbours are : (100,120,110,150,140,130)
         Y_pred = ((100+120+110+150+140+130) / 6

# Feature Scaling

1. Feature scaling is the mothod of scaling numeric features of model in the same scale.
2. Machine learning models dont understand the units(ft, cm, mm, km, etc.) of values. It provides higher
   weightages to largers numeric value and lower weightages to smaller numeric value.
2. So, to avoid the biasness of the data we use feature scaling methods.
3. Feature scaling is the last method in data preprocessing.
4. Feature scaling is used only on the independent variables.
* Following are the techniques used for feature scaling :
  1. MinMax Scaler (Normalization)
  2. Standard Scaler (Standardization)
  3. Robust Scaler
  4. Unit Vector Scaler
  5. Power Transform
  * Out of these methods most commonly used are Normalization and Standardization.
  * In KNN algorithm the most preffered method is Normalization but in some cases Standardization can give
    better accuracy than Normalization.

## 1. MinMax_Scaler (Normalization)

1. This method rescales the feature values in a specific given range.
1. The default range is 0 to 1.
3. Normalized value will be calculated as :
          X_norm = (X - Xmin) / (X_max - X_min)
          where,
          X = Existing value in independent variable.
          X_norm = Normalized value (New)
4. Code for using Normalization :
          normal_scaler = MinMaxScaler()
          x_scaled = normal_scaler.fit_transform(x)
          x_normal = pd.DataFrame(x_scaled, columns=x.columns)
### When to use Normalization :
* When the data is not normally distributed or does not follow Gaussian Distribution then we use
  Normalization technique for scaling.

## 2. Standard_Scaler (Standardization)

1. This method rescales the feature values in random range.
2. Its values dont have any bounding range, it can vary from -3 to +3 or -2.5 to +2.5, etc.
3. Standardized value be given as :
        X_std = (X - mean) / std
        where,
        X = Existing value in independent variable.
        X_std = Standardized Value
        mean = mean of that column
        std = Standard Deviation of that column
4. Code for using Standardization :
        std_scaler = StandardScaler()
        x_scaled = std_scaler.fit_transform(x)
        x_std = pd.DataFrame(x_scaled, columns=x.columns)
        x_std.head()
### When to use Standardization:
* When the data is normally distributed or follow Gaussian Distribution then we use Standardization
  technique for scaling.

# Hyper Parameters

1. Hyper parameters are the parameters which we are used to pass while creating instance of a model.
2. These hyper parameters has specific default values.
3. For KNN default value of k = 5 and P = 2.

# Hyper Parameter tuning
* The process of selecting best value of a hyper parameter for which the model has best accuracy is called
  as "Hyper Parameter tuning".

# Methods of Hyper Parameter Tuning

1. There are two methods used for tuning hyper parameters :
   1. GridSearchCV
   2. RandomizedSearchCV
2. We can use any one of them but generally preffered method is RandomizedSearchCV.
3. In these methods we have to pass the disctonary in which :
       keys = Names of hyper parameter.
       values = range for that hyper parameter.
4. Now this dictionary is to be passed in the instance of GridSearchCV() or RandomizedSearchCV() along
   with model name and the Cross-Validation values (cv = 5).
5. Then we have to train the model using new object.
6. In this way model will be trained on the best value of k (n_neighbours).
7. We can check the best n_neighbours value using code:
       new_object_name.best_estimator_

# Advantages

1. It is very simple algorithm.
2. No assumption about data(It is a Non-Parametric algorithm).
3. It is very useful for non-linear data.
4. Varsetile algorithm (we use for Classsfication and Regression).
5. KNN does not learn any model.

# Disadvantages

1. Lazy Algorithm: (Lazy Learner)
   * All training data used in testing phase.
   * Training is faster and Testing is slower.
2. KNN need more memory.
3. Computation cost is very high due to high memory uses.
4. Not suitable for large amount data(Prediction ).
5. Sensitive to outliers.
6. Sensitive to scale of data(Feature Scaling is required).
7. Always need to determine the value of k.

# Overfitting and Underfitting Logic based on k value

* k = 1 >> Overfitting(Low Bias, High Variance)
* k = 5 >> Best Model(Low Bias , Low Variance)
* k = 100 >> Underfitting(High Bias, Low Variance)
