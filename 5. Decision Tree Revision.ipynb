# Decision Tree 

1. Decision Tree is a Supervised learning technique that can be used for both classification and Regression problems,
   But   mostly it is preferred for solving Classification problems. 
2. In a Decision tree, there are two nodes, which are the Decision Node/branch and Leaf Node. Decision nodes are used 
   to make any decision and have multiple branches, whereas Leaf nodes are the output of those decisions and do not 
   contain any further branches. First Desion node is called as root node.
3. It is called a decision tree because, similar to a tree, it starts with the root node, which expands on further 
   branches and constructs a tree-like structure.

# In order to build a Decision Tree Classification, we use:

  1. ID3 algorithm >> Iterative Dichotomiser 3 >> It uses Information Gain (or entropy) to find best split.
  2. CART algorithm >> Classification and Regression Tree>> It uses Gini Index to find best split. 

# Techniques to select the Best Decision Node

* While implementing a Decision tree, the main issue arises that how to select the best root node and sub-nodes.
* To solve such problems there is a technique/method which is called as Attribute selection measure or ASM.
* There are two popular techniques:
  1. Information Gain
  2. Gini Index

# 1. Information Gain

1. It tells how much information a feature provide us.
2. According to the value of information gain, we split the node and build the decision tree.
3. A decision tree algorithm always tries to maximize the value of information gain, and a node/attribute having 
   the highest information gain is split first. It can be calculated using the below formula:
                 Information Gain = Entropy(S)- [(Weighted Avg) *Entropy(each feature)]


## Entropy

1. It is measure of impurity.
2. Entropy can be calculated as:
                 E(S) = - sum ( P(Yi)log2 P(Yi) )
                 Entropy(s)= -P(yes)log2 P(yes)- P(no) log2 P(no)
                 Where,
                       S= Total number of samples
                       P(yes)= probability of yes
                       P(no)= probability of no

# 2. Gini Index

1. Gini index is a measure of impurity or purity and used while creating a decision tree in the CART
   (Classification and Regression Tree) algorithm.
2. An attribute with the low Gini index should be preferred as compared to the high Gini index.
3. It only creates binary splits, and the CART algorithm uses the Gini index to create binary splits.
4. Gini index can be calculated using the below formula:
                Gini Index = 1 - sum P(Yi)^2

# How does the Decision Tree Classification algorithm Works

### Step-1: 
* The decision tree starts with the root node, which contains the complete dataset.
  (Root node will be decided by Information Gain/Gini Index value of different attributes)
* The attribute which have highest Information Gain value or lowest Gini Index value will be selected as root node or 
  1st decision node. 
  
### Step-2: 
* The root node splits further into the next decision nodes based I.G. or Gini Index values.
  
### Step-3: 
* The next decision nodes further gets split into another decision nodes.

### Step-4: 
* This process will continues untill we get Leaf nodes which are the output of those decisions 
  and do not contain any further branches.

## Example: 

1. Suppose there is a candidate who has a job offer and wants to decide whether he should accept the offer or Not. 
2. So, to solve this problem, the decision tree starts with the root node (Salary attribute decided by 
   Information Gain/Gini Index value).
3. The root node splits further into the next decision node (distance from the office) and one leaf node based on the 
   corresponding labels(output). 
4. The next decision node further gets split into one decision node (Cab facility) and one leaf node. 
5. Finally, the decision node splits into two leaf nodes (Accepted offers and Declined offer).

# Decision Tree Split for Numerical Feature

1. If our feature contains numerical value the 1st thing algorithm will do basically it will be sorting all the 
   values in ascending order.
2. For the particular feature model will find threshold value for which when we split dataset I.G. will be highest 
   or Gini Index will be least.
3. That threshold value will be considered as splitting point.

# Hyper Parameters

#### Hyper Parameters :
  1. Hyper parameters are the parameters which we are used to pass while creating instance of a model.
  2. These hyper parameters has specific default values.

#### Important hyperparameters for Decision Tree Classification:

1. criterion >> 'gini' and 'entropy'
2. min_samples_leaf >> 1 (leaf node)
   * Min number of samples required at leaf node.
3. min_samples_split >> 2  (branch node)
   * Minimum number of samples required to split at branch node.
4. max_depth >> None
   * max_depth : The Maximum depth of tree.

* Overall conclusion : Nodes will expand untill all leaf node contains less than min_samples_split Value.

* Higher max_depth value >> Overfitting
* Lower max_depth value >> Underfitting

# Hyper Parameter tuning
* The process of selecting best value of a hyper parameter for which the model has best accuracy is called
  as "Hyper Parameter tuning".

# Methods of Hyper Parameter Tuning

1. There are two methods used for tuning hyper parameters :
   1. GridSearchCV
   2. RandomizedSearchCV
2. We can use any one of them but generally preffered method is RandomizedSearchCV.
3. In these methods we have to pass the disctonary in which :
       keys = Names of hyper parameter.
       values = range for that hyper parameter.
4. Now this dictionary is to be passed in the instance of GridSearchCV() or RandomizedSearchCV() along
   with model name and the Cross-Validation values (cv = 5).
5. Then we have to train the model using same instance name.
6. In this way model will be trained on the best values of hyperparameters.
7. We can check the best hyperparameters values using code:
       new_object_name.best_params_

# Pruning: Getting an Optimal Decision tree

1. Pruning is a process of deleting the unnecessary nodes from a tree in order to get the optimal decision tree.
2. A too-large tree increases the risk of overfitting, and a small tree may not capture all the important features 
   of the dataset. 
3. Therefore, a technique that decreases the size of the learning tree without reducing accuracy is known as Pruning. 
4. There are mainly two types of tree pruning technology used:
   1. Cost Complexity Pruning (ccp_alpha >> Cost Complexity Pruning)
      * ccp_alpha is one of the hyperparameter for building model.
      * we need to find best value of ccp_alpha for which our model will have low variance and high bais.
   2. Reduced Error Pruning.

# Decision Tree Regression

In case of regression, decision trees learn by splitting the training dataset in a way such that the sum of 
squared error (SSE or residuals) is minimized. It predicts the output value by taking the average of all 
the data points that fall into a leaf node.

# How does the Decision Tree Regression algorithm Works

1. The decision criteria is different for classification and regression trees.
2. The decision criteria in Decision Tree Regression is sum of squared error (SSE or residuals).
3. Decision Tree Regression finds threshold value for which SSE will be least that threshold value will be 
   considered as splitting point.
4. Similarly further branches will also split till we get leaf node. 
5. It predicts the output value by taking the average of all the data points that fall into a leaf node.

# Advantages of the Decision Tree

1. Used for classification and Regression
2. It is non-parametric algorithm
3. Easy to understand, interpret and visualize
4. Best for non-linear relationships dataset
5. Less data preparation is required:
    1. Not impacted by Outlier
    2. Not impacted by Missing Values.

# Disadvantages of the Decision Tree

1. Overfitting: 
   * Low Bias and High Variance
   * Good accuracy on training dataset
   * Not good accuracy on testing dataset
2. It is unstable:
   * Small variation in data might result completely different tree.
3. For more class labels, the computational complexity of the decision tree may increase.

# How to handle Overfitting in decision Tree?

1. Using Ensemble techniques(Random Forest)
2. Pruning (Cutting the trees)
3. By Setting constraints on model parameter.
