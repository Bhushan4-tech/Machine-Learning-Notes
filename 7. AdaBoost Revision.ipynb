# AdaBoost Algorithm

1. It is a Boosting technique used as an Ensemble Method in Machine Learning.
2. It is called Adaptive Boosting as the weights are re-assigned to each instance, with higher 
   weights assigned to incorrectly classified instances. 
3. Boosting is used to reduce bias as well as variance for supervised learning. 
4. It works on the principle of learners growing sequentially. 
5. In simple words, weak learners are converted into strong ones.
6. It creates a Forest of Stumps and Stumps is basically a tree with just one root node and two leaf node.

# Working of AdaBoost Algorithm

1. It makes ‘n’ number of decision trees during the data training period. 
2. As the first decision tree/model is made, the incorrectly classified record in the first model 
   is given priority for next model.
3. In simple words,algorithm converts weak learners to strong ones.
4. The process goes on until we specify a number of base learners we want to create.

### Stump

1. When the random forest is used, the algorithm makes an ‘n’ number of trees. 
2. It makes proper trees that consist of a start node with several leaf nodes. Some trees might be bigger than others, 
   but there is no fixed depth in a random forest. 
3. With AdaBoost, however, the algorithm only makes a node with two leaves, known as Stump.
4. Stumps are weak learners and boosting techniques prefered.

1. All records will be assigned a sample weight. The formula used for this is ‘W=1/N’ where N is the number of records.
2. We can ignore Sample weights(Same) for Creating the First Base Learner.
3. Two properties are considered while selecting a base learner – Gini and Entropy. 
4. We must calculate Gini or Entropy the same way it is calculated for decision trees. 
5. The stump that has the least Entropy or Gini will be selected as the base learner.
6. The 4th is incorrectly classified. For this, we will be calculating the total error.

### Step 5 – Creating a New Dataset

1. In the new dataset, the frequency of incorrectly classified records will be more than the correct ones. 
2. The new dataset has to be created using and considering the normalized weights.
3. To make a new dataset based on normalized weight, the algorithm will divide it into buckets.
4. It will select a random value, see which bucket it is in and select that record for the new dataset. 
5. Based on this new dataset, the algorithm will create a new decision tree/stump and it will repeat the same 
   process till the best value of n_estimators(Number of Stups).
