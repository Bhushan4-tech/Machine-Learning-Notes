# Random Forest

1. Random forest is a Supervised Machine Learning Algorithm that is used widely in Classification and Regression problems.
2. It is based on the concept of ensemble learning, which is a process of combining multiple classifiers to solve a complex  
   problem and to improve the performance of the model.
3. Random Forest is a classifier that contains a number of decision trees on various subsets of the given dataset and 
   final output is considered based on Majority Voting for Classification and Averaging for regression.   

# Ensemble Technique.

* Before understanding the working of the random forest we must look into the ensemble technique. 
* Ensemble simply means combining multiple models.

#### Ensemble uses two types of methods:

1. Bagging (Parallel Approach):
   * It creates a different training subset from sample training data & the final output 
     is based on majority voting. 
   * For example,  
     1. Random Forest.
    
    
2. Boosting(Sequential Approach):
   * It combines weak learners into strong learners by creating sequential models such that the final model 
     has the highest accuracy. 
   * For example,  
     1. AdaBoost (Adaptive Boosting)
     2. Grandient Boosting
     3. XGBoost (Extreme Gradient Boosting).

# Random forest works on the Bagging principle

#### Bagging (Boostrap + Aggregating)

1. Bagging, also known as Bootstrap Aggregation is the ensemble technique used by random forest.
2. Decision Trees are trained in Parallel on various training subset from sample training data & the final output 
   is based on majority voting after combining the results of all models.
3. Bootstrapping Enseures that each individual DT in Random Forest is Unique.
4. This step which involves combining all the results and generating output based on majority voting is known as aggregation.

#### Steps involved in random forest algorithm

* Step-1: 
  * Select random set of data points from the training dataset.

* Step-2: 
  * Build the decision trees associated with the selected data points (Subsets).

* Step 3: 
  * Each decision tree will generate an output.

* Step 4: 
  * Final output is considered based on Majority Voting for Classification and Averaging for regression.

# Hyper Parameters

#### Hyper Parameters :
  1. Hyper parameters are the parameters which we are used to pass while creating instance of a model.
  2. These hyper parameters has specific default values.

#### Important hyperparameters for Decision Tree Classification:

1. n_estimators >> 50 (Decision Trees)
   * number of trees the algorithm builds before averaging the predictions.
2. criterion >> 'gini' and 'entropy' for Classification type and 'mse' for regression type.
3. min_samples_leaf >> 1 (leaf node)
   * Min number of samples required at leaf node.
4. min_samples_split >> 2  (branch node)
   * Minimum number of samples required to split at branch node.
5. max_depth >> None
   * max_depth : The Maximum depth of tree.

* Overall conclusion : Random forest will build with number of decision trees equal to n_estimators and nodes will 
  expand untill all leaf node contains less than min_samples_split Value.

# Important Features of Random Forest

1. Diversity- Not all attributes/variables/features are considered while making an individual tree, each tree is different.

2. Parallelization-Each tree is created independently from different training subset from sample training data. This means
   that we can make full use of the CPU to build random forests.

3. Stability- Stability arises because the result is based on majority voting/ averaging.

# Difference Between Decision Tree & Random Forest

* Decision trees
  1. Decision trees normally suffer from the problem of overfitting if it’s allowed to grow without any control.
  2. A single decision tree is faster in computation.
  3. When a data set with features is taken as input by a decision tree it will formulate some set of rules to do prediction.
  
  
* Random Forest
  1. Random forests are created from subsets of data and the final output is based on average or majority ranking and 
     hence the problem of overfitting is taken care of.
  2. It is comparatively slower.
  3. Random forest randomly selects observations, builds a decision tree and the average result is taken. It doesn’t use 
     any set of formulas.

# Advantages

1. It reduces overfitting in Decision Tree and Helps to improve accuracy
2. Flexible for Classification and Regression
3. It works well for Continuous and categorical data.
4. No Feature scaling required
5. Not impacted by outliers and missing values

# Disadvantages

1. It requires much computation power (Training Multiple Decision Tree)
2. It Also requires much time for training as it combine a multiple DT to determine the class
3. It fails to determine the significance of each variables because each DT will not contain all variables.
