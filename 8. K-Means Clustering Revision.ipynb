# K-Means Clustering Algorithm

1. K-Means Clustering is an unsupervised learning algorithm.
2. It is used to solve the clustering problems in machine learning or data science.
3. It is an iterative algorithm that divides the unlabeled dataset into k different clusters in such a way 
   that each dataset belongs only one group that has similar properties.
4. It is a centroid-based algorithm, where each cluster is associated with a centroid. The main aim of this 
   algorithm is to minimize the sum of distances between the data point and their corresponding clusters.
5. The algorithm takes the unlabeled dataset as input, divides the dataset into k-number of clusters, and repeats 
   the process until it does not find the best clusters. The value of k should be predetermined in this algorithm.   
   
   
#### The k-means clustering algorithm mainly performs two tasks:
1. Determines the best value for K center points or centroids by an iterative process.
2. Assigns each data point to its closest k-center. Those data points which are near to the particular 
   k-center, create a cluster.

# How does the K-Means Algorithm Work?


1. Select the K value to decide the number of clusters.


2. Select random K points or centroids. (It can be other than the input dataset).


3. Use Euclidean Distance to find dist between Centroid point and all data points.


4. Assign each data point to their closest centroid.


5. Update Centroid Value by mean of all datapoints in that particular cluster.
   * Mean = ( Sum of all Datapoints ) / Number of samples 


6. Repeat this process till we get fixed number of data Points in Clusters.

# Elbow Method

1. The Elbow method is used to find the best k value.
2. This method uses the concept of WCSS value.

#### WCSS

1. WCSS stands for Within Cluster Sum of Squares.
2. It is the sum of the square of the distances between each data point and its centroid for each cluster.

3. ∑Pi in Cluster1 distance(Pi C1)2 : It is the sum of the square of the distances between each data point

#### To find the the best k value

2. The point at which abrupt change in curve is considered as the best value of k.
