# Logistic Regression

1. Logistic Regression is Supervised Machine learning classification algorithm that is used to predict the probability of
   categorical variable.
2. It uses Linear regression model so name is regression instead of classification.
3. It is a probabilistic algorithm.


4. Logistic regression produce result in a binary format which is used to predict the outcome of categorical
   dependent variable. So the outcome should be discrete/categorical such as: 0 or 1, yes or No, True or
   False etc.

## Types of Logistic Regression

1) Binomial: target variable can have only 2 possible types: “0” or “1” which may represent “win” vs “loss”, “pass”
   vs “fail” etc.
   
2) Multinomial: target variable can have 3 or more possible types which are not ordered(i.e. types have no
   quantitative significance).
   
3) Ordinal: it deals with target variables with ordered categories. For example, a test score can be categorized
   as:“very poor”, “poor”, “good”, “very good”. Here, each category can be given a score like 0, 1, 2, 3.

# Function Used in Logistic Regression

1. Function used in Logistic Regression is "Sigmoid Fuction".
   * Logistic Regression Function(Sigmoid Fuction):
         P(y=1) = 1/(1 + e^-y)
         P(y=1) = 1/(1 + e^-(mx + c))
2. The sigmoid function also called a logistic function.
3. It is a mathematical function having a characteristic that can take any real value and map it to between 0 to 1.


4. If the value of y goes to positive infinity then the predicted value of y will become 1 and if it goes to
   negative infinity then the predicted value of y will become 0.
5. let's say we have probability of getting y = 1 is 0.8 {i.e. p(y=1) = 0.8}, So in this case how we can 
   decide whether our value 0 or 1, for that we have "Threshold" value.
6. If value greater than threshold value then it is consider as 1 and if it is less than thresold value 
   then it is consider as 0.

# Logit Function

1. When we do log on both side of Logistic function and after solving will get Logit odd function.
2. It also known as log odd function, that is nothing but log of the probability of an event happening (P) / the 
   probability of an event not happening (1-P).
   * Formula:
            Y=Log(P/1-P)
            log(P/(1-P)) = y = m1x1 + m2x2 +c

# log odds

* The odds ratio is the probability of success/probability of failure. As an
  equation, that’s P(A)/P(-A), where P(A) is the probability of A, and P(-A) the
  probability of not A.
  
* Taking the logarithm of the odds ratio gives us the log odds of A, which can be
  written as

  * log(A) = log(P(A)/P(-A))

* Since the probability of an event happening, P(-A) is equal to the probability of
  an event not happening, 1 – P(A), we can write the log odds as

  * log [p/(1-p)]
    * p = the probability of an event happening
    * 1 – p = the probability of an event not happening

## Cost Function or LogLoss Function

1. In Logistic Regression LogLoss Function is used as cost function.
2. This function measures the performance of classification model.
3. LogLoss increases as the predicted probability diverges from actual value.
4. Formula:
          LogLoss Function = -1/N sum(Yi log(Pi) + (1-Yi) log(1-Pi))
          where, N = Number of data points
                 Yi = Actual value of y at point i (either 0 or 1)
                 Pi = probability (predicted value of y) at point i (value ranges from 0 to 1)

* IF yi = 0 , then LogLoss Function will become, LL = -log(1-pi) ,  it will give exponentially increasing curve.
* If yi = 1, then LogLoss Function will become, LL = -log(pi) ,  it will give exponentially decreasing curve.

## Why MSE is not used as a cost function in Logistic Regression

1. If we use MSE as cost function we will get non convex curve.
2. The nonconvex curve consists of number of local minima points, and it is very diffucult to get exact gobal
   minima point.
3. One more reason is we have target values like 0 or 1 and predicted value will be in 0 to 1 range, So MSE will always be in      between 0-1, which can make it very difficult to keep track of the errors.

##  Confusion Matrix

1. The confusion matrix is a matrix used to determine the performance of the classification
   models for a given set of test data.
2. It can only be determined if the true values for test data are known.
                                1 0 <- Actual Values.
         Confusion Matrix => 1 TP FP
         Predicted Values -> 0 FN TN

#### Confusion Matrix Components:

1. True Positive-
   * When predicted is true i.e. positive, so actual is also positive.
   * When both actual and predicted values are in class 1.
2. False Negative-
   * When predicted is false i.e. negative, but actual is positive.
   * When predicted class 0 but actaul in class 1.
3. True Negative-
   * When predicted is true i.e. negative, so actual is also negative.
   * When both actual and predicted are in class 0.
4. False Positive-
   * When predicted is false i.e. positive, but actual is negative.
   * When predicted in class 1 but actal in class 0.

#### It includes following parameters:

1. TPR/Recall/Sensitivity(True Positive Rate):
   * It gives the proportion of positive class correctly classified by classifier.
   * TPR = TP/(TP+FN)
      
      
2. FNR(False Negative Rate):
   * It gives the proprotion of positive class got incorrectly classified by classifier.
   * FNR = FN/(TP+FN)
   
   
3. TNR/Specificity(True Negative Rate):
   * It gives the proportion of negative class correctly classified classifier.
   * TNR = TN/(TN+FP)
   
   
4. FPR(Flase Positive Rate):
   * It gives the proportion of nagative class got incorrectly classified by classifier. 
   * FPR = FP/(FP+TN) = 1 - TNR = 1 - Specificity
       
       
* Conclussions:
  * For class 1 TPR should be high and FNR shoud low.
  * For class 0 TNR shoud be high and FPR low.

#### Error Types:

1. Type 1 Error:
   * FP is known as type 1 error.
   * This error belongs to negative class or class 0 where actual value is 0 but predicted 1.
2. Type 2 Error:
   * FN is known as type 2 error.
   * This error belongs to positive class or class 1 where actual value is 1 but predicted 0.

## Classification Report

1. Recall/Sensitivity/TPR:
   * It gives the proportion of positive class correctly classified by classifier.
   * Out of total actual positive result (1),how many are positive predicted results.
   * When FN has higher importance.
   * Recall = TP/(TP+FN)
   * Eg-
      * Cancer Disease Predictions (must be Low FN values).
      * Yes or No by using various feature(Medical Reports)
      * If a person Having a cancer but he predicted as No.(this statement is very much important)


2. Precision :
   * Out of total predicted positive results, how many results are actual positive.
   * When FP has higher importance.
   * Precision = TP/(TP+FP)
   * Eg-Spam Filter


3. F1-Score/F-Beta Score:
   * When Precision and Recall both are important.
   * When FP & FN both are important or No clear distinction between Precision & recall.
      * FB-Score = (1+B^2)P*R/(B^2*P+R)
   * When Beta eqals to 1 FB-Score becomes F1-Score i.e. Harmonic mean of Precision and Recall value.
   * F1-Score = 2PR/P+R
   * Eg: Titanic passengers dataset,when survieved and not survived both are important.

## ROC Curve:

* ROC-Receiver Operating Characteristics:
  * The Receiver Operator Characteristic (ROC) curve is an evaluation metric for binary
    classification problems.
  * ROC curve help us to visualize how well our Machine Learning classifier is performing. It works only for
    binary classification problems.
  * ROC is the graphical representation of TPR vs FPR or (Sensitivity vs 1-Specificity) at various threshold
    values.
   where,
   * TPR = TP/(TP+FN)
   * FPR = FP/(FP+TN) = 1 - TNR = 1 - Specificity


* AUC-Area Under Curve:
  * Area Under Curve is measure of the, ability of a classifier to distinguish between the positive and negative classes. 
  * The higher the AUC, the better the performance of model distinguishing between the positive and negative classes.

# Bias and Variance

1. Bais : Bias is the amount that, how much our model’s prediction differs from the target value, for the training data.

   1. Training Dataset:
      1. Linear Model >> Sum of Squares (Yi - Yp)>> 256 (i.e. large value) >> High Bias
      2. Polynomial Regression Model >> Sum of Squares (Yi - Yp) >> 0 (i.e. low value) >> Low Bias
         (DT , KNN)

   2. Testing Dataset: 
      1. Linear Model >> Sum of Squares (Yi - Yp) >> 250 - 270 >> 
      2. Polynomial Regression Model >> Sum of Squares (Yi - Yp) >> 350 >>


2. Variance : The Difference between the accuracies of different datasets(Train, Test).

       High Variance : Training Data Accuracy >> High
                       Testing Data Accuracy  >> Low
            
       Low Variance: Training Data Accuracy >> High
                     Testing Data Accuracy  >> High
                    
       Low Variance: Training Data Accuracy >> Low
                     Testing Data Accuracy  >> Low

 
* Overfitting  >> Low Bias and High Variance
* Underfitting >> High Bias and Low Variance
* Ideal Model  >> Low Bias and Low Variance

# Difference between Linear Regression and Logistic Regression

* Linear Regression:
  1. Linear Regression is used when our target variable is continuos values. For eg.
     price,age height,weight.
  2. Linear Regression uses the ordinary least squre method to minimize the error and arrives
     at the best possible solution.
  3. Gradient descent algorithm used to find best fit line.   
  4. Cost function is MSE.
  5. In evalution part we are finding MSE,r2_score.


* Logistic Regression:
  1. Logistic Regression is used when the target variable is binary or limited classification
     For eg . yes or no, True or false.
  2. Logistic regression uses probabilistic algorithm.
  3. Gradient descent algorithm used to find best sigmoid curve.
  4. Cost function is Logloss function. 
  5. In evalution part we are finding accuracy score,confusion matrix,classification report,ROC curve.
